7. Selecting different features from step 6 implement a linear regression algorithm and find the slope, the intercept, and the error of the regression model.
8. Display the line of best fit from step 7.
9. LinearRegression()fromvsklearn


```python
from sklearn.linear_model import LinearRegression

X = np.array(data1['median_income']).reshape(-1, 1)
y = np.array(data1['median_house_value']).reshape(-1, 1)

LR = LinearRegression()
LR.fit(X, y)
print("Slope: ", LR.coef_)
print("Intercept: ", LR.intercept_)
print("Error:", (1 - LR.score(X, y)))
y_pred = LR.predict(X)
plt.scatter(data1['median_income'], data1['median_house_value'], s = 3)
plt.plot(X, y_pred, color = 'r')
plt.show()
```
